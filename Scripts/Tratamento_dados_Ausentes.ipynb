### P059

```python
data[['P050|P052','P05901', 'P05902', 'P05903', 'P05904']]
```

```python
data[['P05901', 'P05902', 'P05903', 'P05904']].info()
```

```python
data['P05904'] = data['P05904'] / 365
data['P05903'] = data['P05903'] / 52
data['P05902'] = data['P05902'] / 12
```

```python
data['P05901'] += data['P05902'] + data['P05903'] + data['P05904']
```

```python
data['P05901'].max()
```

```python
bins = [0, 0.99, 5, 10, 20, 47]
labels = ['Menos de 1 ano', '1-5', '6-10', '11-20', '21+']

data['P05901'] = pd.cut(data['P05901'], bins=bins, labels=labels, right=False)
```

```python
data.drop(columns=['P05902', 'P05903', 'P05904'], inplace=True)
```

```python
temp = data.loc[data['P050|P052'] == 1].copy() 

if pd.api.types.is_categorical_dtype(temp['P05901']):
    temp['P05901'] = temp['P05901'].cat.add_categories(['Fuma atualmente'])
    data['P05901'] = data['P05901'].cat.add_categories(['Fuma atualmente'])

data.loc[data['P050|P052'] == 1, 'P05901'] = temp['P05901'].fillna('Fuma atualmente')
```

```python
temp = data.loc[data['P050|P052'] == 2].copy() 

data.loc[data['P050|P052'] == 2, 'P05901'] = temp['P05901'].fillna('Fuma atualmente')
```

```python
temp = data.loc[data['P050|P052'] == 4].copy() 

if pd.api.types.is_categorical_dtype(temp['P05901']):
    temp['P05901'] = temp['P05901'].cat.add_categories(['Nunca fumou'])
    data['P05901'] = data['P05901'].cat.add_categories(['Nunca fumou'])

data.loc[data['P050|P052'] == 4, 'P05901'] = temp['P05901'].fillna('Nunca fumou')
```
