```python
import pandas as pd
from sklearn.impute import KNNImputer
```

### P00104

```python
base['P00104'].isna().sum()
```

```python
coluna_alvo = base[['P00104']]

imputer = KNNImputer(n_neighbors=4)
coluna_imputada = imputer.fit_transform(coluna_alvo)
base['P00104'] = coluna_imputada
```
