```python
import pandas as pd
from sklearn.impute import KNNImputer
```

```python
base = pd.read_excel("../../Bases/Base_plano_Cartesiano.xlsx")
```

```python
base
```

```python
base.info()
```

### Separacao das classes

```python
doente = base.loc[base['Q074'] == 1]
n_doente = base.loc[base['Q074'] == 2]
```

### P00104

#### Doente

```python
doente['P00104'].describe()
```

```python
doente['P00104'].isna().sum()
```

```python
coluna_alvo = doente[['P00104']]

imputer = KNNImputer(n_neighbors=4)
coluna_imputada = imputer.fit_transform(coluna_alvo)
doente['P00104'] = coluna_imputada
```

```python
doente['P00104'].isna().sum()
```

```python
doente['P00104'].describe()
```

#### Nao Doente

```python
print(n_doente['P00104'].describe())
print(n_doente['P00104'].isna().sum())
```

```python
coluna_alvo = n_doente[['P00104']]

imputer = KNNImputer(n_neighbors=4)
coluna_imputada = imputer.fit_transform(coluna_alvo)
n_doente['P00104'] = coluna_imputada
```

```python
print(n_doente['P00104'].describe())
print(n_doente['P00104'].isna().sum())
```
